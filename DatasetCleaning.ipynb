# **Dataset Merging Using SQL (Added 13 January, 2026)**

```python
# ===== SQL MERGE CELL (DuckDB) - AUTO MAPPING, NO PLACEHOLDERS =====
!pip -q install duckdb

import duckdb
import pandas as pd

con = duckdb.connect()

# Register your dataframes (use your cleaned/mined dfs here if they have different names)
con.register("mhcc", df1)
con.register("mc16k", df2)
con.register("prosocial", df3)
con.register("empathetic_dialogues", df4)
con.register("ethics_commonsense", df5)
con.register("ethics_deontology", df6)
con.register("ethics_justice", df7)
con.register("amazon_commonsense_dialogues", df8)

tables_sources = [
    ("mhcc", "Amod/mental_health_counseling_conversations"),
    ("mc16k", "ShenLab/MentalChat16K"),
    ("prosocial", "allenai/prosocial-dialog"),
    ("empathetic_dialogues", "kaggle:empathetic-dialogues-facebook-ai"),
    ("ethics_commonsense", "lighteval/hendrycks_ethics:commonsense"),
    ("ethics_deontology", "lighteval/hendrycks_ethics:deontology"),
    ("ethics_justice", "lighteval/hendrycks_ethics:justice"),
    ("amazon_commonsense_dialogues", "gretelai/commonsense-dialogues"),
]

# Heuristics for common prompt/response column names
PROMPT_CANDIDATES = [
    "prompt","instruction","input","question","query","context","text","dialogue","utterance","human","user","turns"
]
RESPONSE_CANDIDATES = [
    "response","output","answer","assistant","bot","completion","reply","label","chosen","assistant_response"
]

def pick_column(cols, candidates):
    cols_lower = {c.lower(): c for c in cols}
    for cand in candidates:
        if cand in cols_lower:
            return cols_lower[cand]
    return None

union_parts = []
mapping_report = {}

for table, source in tables_sources:
    cols = con.execute(f"DESCRIBE {table}").df()["column_name"].tolist()
    pcol = pick_column(cols, PROMPT_CANDIDATES)
    rcol = pick_column(cols, RESPONSE_CANDIDATES)

    # If we can’t confidently map, we skip that table and report it
    if not pcol or not rcol:
        mapping_report[table] = {"source": source, "prompt_col": pcol, "response_col": rcol, "cols": cols}
        continue

    mapping_report[table] = {"source": source, "prompt_col": pcol, "response_col": rcol}

    union_parts.append(f"""
        SELECT
            CAST("{pcol}" AS VARCHAR) AS p,
            CAST("{rcol}" AS VARCHAR) AS r,
            '{source}' AS source
        FROM {table}
    """)

# Show what got mapped / what got skipped
mapping_report
```

```python
if not union_parts:
    raise ValueError("No tables could be auto-mapped. Check mapping_report for column names and map manually.")

sql = f"""
WITH unified AS (
    {'UNION ALL'.join(union_parts)}
),
cleaned AS (
    SELECT
        TRIM(REGEXP_REPLACE(p, '\\\\s+', ' ')) AS prompt,
        TRIM(REGEXP_REPLACE(r, '\\\\s+', ' ')) AS response,
        source
    FROM unified
    WHERE p IS NOT NULL AND r IS NOT NULL
      AND LENGTH(TRIM(p)) > 0
      AND LENGTH(TRIM(r)) > 0
),
dedup AS (
    SELECT prompt, response, source
    FROM (
        SELECT
            prompt,
            response,
            source,
            ROW_NUMBER() OVER (PARTITION BY prompt, response ORDER BY source) AS rn
        FROM cleaned
    )
    WHERE rn = 1
)
SELECT * FROM dedup
"""

merged_df = con.execute(sql).df()
merged_df.head()

merged_df.info()

```

```python
"""
Dataset Merger - Combines multiple mental health & ethics datasets
Exports to CSV, JSON, and Parquet formats for training
"""

import pandas as pd
from datasets import load_dataset
import kagglehub
import os
import json

def load_all_datasets():
    """Load all datasets from various sources"""
    print("Loading datasets...")

    # Load HuggingFace datasets
    print("1/8 Loading Mental Health Counseling Conversations...")
    dsMHCC = load_dataset("Amod/mental_health_counseling_conversations")
    df1 = pd.DataFrame(dsMHCC['train'])

    print("2/8 Loading MentalChat16K...")
    dsMC16k = load_dataset("ShenLab/MentalChat16K")
    df2 = pd.DataFrame(dsMC16k['train'])

    print("3/8 Loading Prosocial Dialog...")
    dsPD = load_dataset("allenai/prosocial-dialog")
    df3 = pd.DataFrame(dsPD['train'])

    print("4/8 Loading Empathetic Dialogues (Kaggle)...")
    try:
        path = kagglehub.dataset_download("atharvjairath/empathetic-dialogues-facebook-ai")
        csv_path = None
        for file in os.listdir(path):
            if file.endswith(".csv"):
                csv_path = os.path.join(path, file)
                break
        df4 = pd.read_csv(csv_path) if csv_path else pd.DataFrame()
    except Exception as e:
        print(f"Warning: Could not load Kaggle dataset: {e}")
        df4 = pd.DataFrame()

    print("5/8 Loading Ethics - Commonsense...")
    dsECMS = load_dataset("lighteval/hendrycks_ethics", "commonsense")
    df5 = pd.DataFrame(dsECMS['train'])

    print("6/8 Loading Ethics - Deontology...")
    dsDY = load_dataset("lighteval/hendrycks_ethics", "deontology")
    df6 = pd.DataFrame(dsDY['train'])

    print("7/8 Loading Ethics - Justice...")
    dsJ = load_dataset("lighteval/hendrycks_ethics", "justice")
    df7 = pd.DataFrame(dsJ['train'])

    print("8/8 Loading Commonsense Dialogues...")
    dsACMS = load_dataset("gretelai/commonsense-dialogues")
    df8 = pd.DataFrame(dsACMS['train'])

    return [df1, df2, df3, df4, df5, df6, df7, df8]

def _safe_cell_to_str(x):
    """Convert any value to a Parquet-safe string without losing structured info."""
    if x is None:
        return ""
    # pandas NA / NaN
    try:
        if pd.isna(x):
            return ""
    except Exception:
        pass

    # lists/dicts/tuples/sets -> JSON string
    if isinstance(x, (list, dict, tuple, set)):
        try:
            return json.dumps(x, ensure_ascii=False)
        except Exception:
            return str(x)

    return str(x)

def standardize_format(dfs):
    """Standardize all datasets to have consistent columns: prompt, response, source"""
    standardized = []

    for i, df in enumerate(dfs, 1):
        if df.empty:
            continue

        print(f"\nProcessing dataset {i}...")
        print(f"Original columns: {df.columns.tolist()}")
        print(f"Shape: {df.shape}")

        # Create standardized dataframe
        std_df = pd.DataFrame()

        # Determine source based on dataset index or existing column
        if 'source' in df.columns:
            source = df['source'].iloc[0] if len(df) > 0 else f"dataset_{i}"
        else:
            sources = [
                "Amod/mental_health_counseling_conversations",
                "ShenLab/MentalChat16K",
                "allenai/prosocial-dialog",
                "empathetic-dialogues-facebook-ai",
                "lighteval/hendrycks_ethics-commonsense",
                "lighteval/hendrycks_ethics-deontology",
                "lighteval/hendrycks_ethics-justice",
                "gretelai/commonsense-dialogues"
            ]
            source = sources[i-1] if i <= len(sources) else f"dataset_{i}"

        # --- Dataset-specific mapping fix (prevents grabbing Unnamed: 0 as prompt) ---
        # Kaggle Empathetic Dialogues: use Situation -> empathetic_dialogues
        if i == 4 and ("Situation" in df.columns) and ("empathetic_dialogues" in df.columns):
            std_df["prompt"] = df["Situation"]
            std_df["response"] = df["empathetic_dialogues"]

        else:
            # Map columns to standard format
            # Try to find prompt/response columns with common names
            prompt_cols = ['prompt', 'question', 'input', 'text', 'Context', 'context', 'instruction', 'Situation', 'scenario']
            response_cols = ['response', 'answer', 'output', 'completion', 'utterance', 'Response', 'excuse', 'empathetic_dialogues', 'turns']

            prompt_col = next((col for col in prompt_cols if col in df.columns), None)
            response_col = next((col for col in response_cols if col in df.columns), None)

            if prompt_col and response_col:
                std_df['prompt'] = df[prompt_col]
                std_df['response'] = df[response_col]

            else:
                # --- Smarter fallback: avoid ID/label/Unnamed columns and prefer text-ish columns ---
                bad_substrings = ("unnamed", "id", "label", "group", "index")
                candidate_cols = []
                for col in df.columns:
                    name = str(col).lower()
                    if any(b in name for b in bad_substrings):
                        continue
                    candidate_cols.append(col)

                # Prefer object/string-like cols first
                textish_cols = [c for c in candidate_cols if df[c].dtype == "object"]
                use_cols = textish_cols if len(textish_cols) >= 2 else candidate_cols

                if len(use_cols) >= 2:
                    std_df['prompt'] = df[use_cols[0]]
                    std_df['response'] = df[use_cols[1]]
                else:
                    print(f"Warning: Could not standardize dataset {i}, skipping...")
                    continue

        std_df['source'] = source

        # Remove any rows with null values
        std_df = std_df.dropna()

        print(f"Standardized shape: {std_df.shape}")
        standardized.append(std_df)

    return standardized

def merge_and_export(dfs, output_dir="/content/outputs"):
    """Merge all datasets and export to multiple formats"""

    # Merge all dataframes
    print("\nMerging all datasets...")
    merged_df = pd.concat(dfs, ignore_index=True)

    # --- Parquet safety: force consistent, serializable types ---
    # This prevents ArrowTypeError when prompt/response contain ints/lists/dicts/etc.
    for col in ["prompt", "response", "source"]:
        merged_df[col] = merged_df[col].map(_safe_cell_to_str)

    print(f"\n{'='*60}")
    print(f"MERGED DATASET SUMMARY")
    print(f"{'='*60}")
    print(f"Total rows: {len(merged_df):,}")
    print(f"Total columns: {len(merged_df.columns)}")
    print(f"\nRows per source:")
    print(merged_df['source'].value_counts())
    print(f"\nMemory usage: {merged_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    print(f"{'='*60}\n")

    # Create output directory
    os.makedirs(output_dir, exist_ok=True)

    # Export to multiple formats
    print("Exporting datasets...")

    # 1. CSV (most compatible)
    csv_path = os.path.join(output_dir, "merged_training_dataset.csv")
    merged_df.to_csv(csv_path, index=False)
    print(f"✓ Exported to CSV: {csv_path}")

    # 2. Parquet (most efficient for large datasets)
    parquet_path = os.path.join(output_dir, "merged_training_dataset.parquet")
    merged_df.to_parquet(parquet_path, index=False)
    print(f"✓ Exported to Parquet: {parquet_path}")

    # 3. JSON Lines (good for streaming)
    jsonl_path = os.path.join(output_dir, "merged_training_dataset.jsonl")
    merged_df.to_json(jsonl_path, orient='records', lines=True, force_ascii=False)
    print(f"✓ Exported to JSONL: {jsonl_path}")

    # 4. Sample for preview
    sample_path = os.path.join(output_dir, "sample_100_rows.csv")
    merged_df.head(100).to_csv(sample_path, index=False)
    print(f"✓ Exported sample: {sample_path}")

    return merged_df, [csv_path, parquet_path, jsonl_path, sample_path]

if __name__ == "__main__":
    # Load datasets
    dfs = load_all_datasets()

    # Standardize format
    standardized_dfs = standardize_format(dfs)

    # Merge and export
    merged_df, file_paths = merge_and_export(standardized_dfs)

    print("\n✅ All done! Your training dataset is ready.")
    print(f"\nFiles created:")
    for path in file_paths:
        print(f"  - {path}")
```
